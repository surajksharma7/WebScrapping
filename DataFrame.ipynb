# Forming Data Frame 

```python
import numpy as np #importing numppy
import pandas as pd # importing pandas
from bs4 import BeautifulSoup # importing beautifulsoup
```

```python
# Opening file from html format in readable format
with open('smartphone.html','r',encoding='utf-8') as f:
    html = f.read() 
```

```python
soup = BeautifulSoup(html,'lxml') 
```

```python
container = soup.find_all('div',{'class' : 'sm-product has-tag has-features has-actions'})
```

```python
# arranging data in topic wise so that it can form dataset

name = []
price = []
specs = []
sim = []
processor = []
storage = []
battery = []
display = []
camera = []
ram = []
version = []

for i in container:
    
    try:
        name.append(i.find('h2').text)
    except:
        name.append(np.nan)
        
    try:
        price.append(i.find('span',{'class':'price'}).text)
    except:
        price.append(np.nan)
        
    try:
        specs.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        specs.append(np.nan)

    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    try:
        sim.append(x[0].text)
    except:
        sim.appendd(np.nan)
        
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
        
        
    try:
        storage.append(x[2].text)
    except:
        storage.append(np.nan)

    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
        
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
        
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
        
    try:
        ram.append(x[6].text)
    except:
        ram.append(np.nan)

    try:
        version.append(x[7].text)
    except:
        version.append(np.nan)
```

```python
# Creating Dataframe from list

smartphone = pd.DataFrame({
    'model' : name,
    'price' : price,
    'rating' : specs,
    'name': name,
    'price':price,
    'specs':specs,
    'sim':sim,
    'processor':processor,
    'storage': storage,
    'battery':battery,
    'display': display,
    'camera': camera,
    'ram':ram,
    'version':version
})
```
